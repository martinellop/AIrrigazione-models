```python
import torch
import os
from model import skyModel
from model import skyDataset
import random
import numpy as np
import cv2
import matplotlib.pyplot as plt
from torch.utils.data import DataLoader
from torch.utils.data import Dataset
```

```python
model = skyModel()
model.load_state_dict(torch.load("skyModel.bin"))
model.eval()
```

```python
parentDir = os.path.join(".","original_pictures")
_, dirs, _ = next(os.walk(parentDir))
category = os.path.join(parentDir, random.choice(dirs))
pic = os.path.join(category,random.choice(os.listdir(category)))
print(pic)
img = cv2.imread(pic)
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img)
plt.plot()

result = model.evaluatePicture(img)
sentence = "it will rain." if result else "no, it will not rain."
print("result: ",sentence)
```
